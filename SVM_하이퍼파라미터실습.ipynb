# 붓꽃 데이터셋을 사용하여, 특성 스케일을 변경하고, Iris-Virginia 품종을 감지하기 위해 사이킷런의 선형 SVM 모델을 훈련 
# 이 때,  하이퍼파라메터 C가 1, 50, 100일 때의 산점도 그래프를 각각 출력하기

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline

iris = datasets.load_iris()

X = iris["data"][:,(2,3)]
y = (iris['target']==2).astype(np.float64)
```

```python
#C=1로 모델 훈련 및 산점도 출력

svm_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_svc", LinearSVC(C=1, loss="hinge", max_iter=50000)),
])
svm_clf.fit(X, y)

w=-svm_clf.named_steps["linear_svc"].coef_[0,0]/svm_clf.named_steps["linear_svc"].coef_[0,1]
b=-svm_clf.named_steps["linear_svc"].intercept_[0]/svm_clf.named_steps["linear_svc"].coef_[0,1]

line =svm_clf.named_steps["scaler"].inverse_transform([[-5, (-5*w)+b],[5,5*w+b]])

plt.figure()
plt.plot(line[:,0], line[:,1], 'r:', label="LinearSVC")
plt.plot(X[:,0][y==1], X[:,1][y==1], 'bo', label="Iris-Virginica")
plt.plot(X[:,0][y==0], X[:,1][y==0], 'yo', label="Not-Virginica")

plt.xlabel("petal length", fontsize=14)
plt.ylabel("petal width", fontsize=14)
plt.axis([0,7,0,2.5])
plt.show()
```

```python
#C=50으로 모델 훈련 및 산점도 출력

svm_clf_50 = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_svc", LinearSVC(C=50, loss="hinge", max_iter=50000)),
])
svm_clf_50.fit(X, y)

w_50=-svm_clf_50.named_steps["linear_svc"].coef_[0,0]/svm_clf_50.named_steps["linear_svc"].coef_[0,1]
b_50=-svm_clf_50.named_steps["linear_svc"].intercept_[0]/svm_clf_50.named_steps["linear_svc"].coef_[0,1]

line_50 =svm_clf_50.named_steps["scaler"].inverse_transform([[-5, (-5*w_50)+b_50],[5,5*w_50+b_50]])

plt.figure()
plt.plot(line_50[:,0], line_50[:,1], 'r:', label="LinearSVC")
plt.plot(X[:,0][y==1], X[:,1][y==1], 'bo', label="Iris-Virginica")
plt.plot(X[:,0][y==0], X[:,1][y==0], 'yo', label="Not-Virginica")

plt.xlabel("petal length", fontsize=14)
plt.ylabel("petal width", fontsize=14)
plt.axis([0,7,0,2.5])
plt.show()
```
